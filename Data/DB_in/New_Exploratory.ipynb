```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("CORN_in.csv")
```

```python
data.head()
```

```python
data['SP500_Vol'].plot(color = 'k', figsize = (20,3))
plt.title("SP500 Scaled Intraday Range")
plt.show()
```

```python
data['BCOMAG_Vol'].plot(color = 'k', figsize = (20,3))
plt.title("Bloomberg Agricultural Subindex Scaled Intraday Range")
plt.show()
```

```python
data['BCOMEN_Vol'].plot(color = 'k', figsize = (20,3))
plt.title("Bloomberg Energy Subindex Intraday Range")
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
plt.scatter(data["SP500_Vol"], data['BCOMEN_Vol'], color = "k", s = 3)
plt.xlabel("SP500 Scaled Intraday Range")
plt.ylabel("Bloomberg Energy Subsindex Scaled Intraday Range")
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
plt.scatter(data["SP500_Vol"], data['BCOMAG_Vol'], color = "k", s = 3)
plt.xlabel("SP500 Scaled Intraday Range")
plt.ylabel("Bloomberg Agriculture Subsindex Scaled Intraday Range")
plt.show()
```
